### Section 1 - Customer Spending

#### 1. Top 10 Customers by Total Spend

```python
query = """select
                customer_id,
                count(*) as total_orders,
                round(sum(cost_of_the_order),2) as total_spend
           from product_events
           group by customer_id
           order by total_spend desc
           limit 10"""
pd.read_sql(query, engine)
```

#### 2. Average Spend Per Customer

```python
query = """select
                customer_id,
                round(avg(cost_of_the_order), 2) as avg_order_value
           from product_events
           group by customer_id
           order by avg_order_value desc"""
pd.read_sql(query, engine)
```

#### 3. Customers With Highest Number of Orders

```python
query = """select
                customer_id,
                count(*) as total_orders
           from product_events
           group by customer_id
           order by total_orders desc"""
pd.read_sql(query, engine)
```

#### 4. Customer Segmentation

```python
query = """select
                customer_id,
                round(sum(cost_of_the_order), 2) as total_spend,
                case
                    when sum(cost_of_the_order) >= 100 then 'high value'
                    when sum(cost_of_the_order) >= 50 then 'medium value'
                else 'low value'
                end as customer_segment
           from product_events
           group by customer_id
           order by total_spend desc"""
pd.read_sql(query, engine)
```

#### 5. Average Restaurants Viewed

```python
query = """select
                round(avg(restaurants_viewed), 2) as avg_restaurants_viewed
           from product_events"""
pd.read_sql(query, engine)
```

#### 6. Customers Viewing The Most Restaurants

```python
query = """select
                customer_id,
                sum(restaurants_viewed) as total_views
           from product_events
           group by customer_id
           order by total_views desc
           limit 10"""
pd.read_sql(query, engine)
```

#### 7. Coupon Usage Percentage

```python
query = """select
                round(avg(coupon_used) * 100, 2) as coupon_usage_percentage
           from product_events"""
pd.read_sql(query, engine)
```

#### 8. Coupon Usage by City

```python
query = """select
                city,
                round(avg(coupon_used) * 100, 2) as coupon_usage_percentage
           from product_events
           group by city
           order by coupon_usage_percentage desc"""

pd.read_sql(query, engine)
```

#### 9. Average Order Value by Acquisition Channel

```python
query = """
select
    acquisition_channel,
    round(avg(cost_of_the_order), 2) as avg_order_value
from product_events
group by acquisition_channel
order by avg_order_value desc
"""

pd.read_sql(query, engine)
```

#### 10. Customer Count by Acquisition Channel

```python
query = """
select
    acquisition_channel,
    count(distinct customer_id) as total_customers
from product_events
group by acquisition_channel
order by total_customers desc
"""

pd.read_sql(query, engine)
```

#### 11. Revenue by Device Type

```python
query = """
select
    device_type,
    round(sum(cost_of_the_order), 2) as total_revenue
from product_events
group by device_type
order by total_revenue desc
"""

pd.read_sql(query, engine)
```

#### 12. Average Rating by Device Type

```python
query = """
select
    device_type,
    round(avg(rating), 2) as avg_rating
from product_events
group by device_type
order by avg_rating desc
"""

pd.read_sql(query, engine)
```

#### 13. Recommendation Click-Through Rate (CTR)

```python
query = """
select
    round(avg(recommendation_clicked) * 100, 2) as recommendation_ctr
from product_events
"""

pd.read_sql(query, engine)
```

#### 14. Recommendation CTR by Device

```python
query = """
select
    device_type,
    round(avg(recommendation_clicked) * 100, 2) as recommendation_ctr
from product_events
group by device_type
order by recommendation_ctr desc
"""

pd.read_sql(query, engine)
```

#### 15. Cart Creation Rate by Acquisition Channel

```python
query = """
select
    acquisition_channel,
    round(avg(cart_created) * 100, 2) as cart_creation_rate
from product_events
group by acquisition_channel
order by cart_creation_rate desc
"""

pd.read_sql(query, engine)
```

#### 16. Rank Customers by Total Spend

```python
query = """
select
    customer_id,
    round(sum(cost_of_the_order), 2) as total_spend,
    rank() over (
        order by sum(cost_of_the_order) desc
    ) as spending_rank
from product_events
group by customer_id
"""

pd.read_sql(query, engine)
```

#### 17. Dense Rank Customers by Total Spend

```python
query = """
select
    customer_id,
    round(sum(cost_of_the_order), 2) as total_spend,
    dense_rank() over (
        order by sum(cost_of_the_order) desc
    ) as spending_rank
from product_events
group by customer_id
"""

pd.read_sql(query, engine)
```

#### 18. Row Number by Customer Spend

```python
query = """
select
    customer_id,
    round(sum(cost_of_the_order), 2) as total_spend,
    row_number() over (
        order by sum(cost_of_the_order) desc
    ) as row_num
from product_events
group by customer_id
"""

pd.read_sql(query, engine)
```
