```python
import os

password = os.environ["DB_PASSWORD"]
```

```python
from sqlalchemy import create_engine

engine = create_engine(
                        f"mysql+pymysql://root:{password}@localhost:3306/foodhub"
                        )

print("Connection successful!")
```

### SECTION 1: Revenue Analysis

#### 1. Total Revenue

```python
query = """select
                round(sum(cost_of_the_order),2) as total_revenue
           from product_events"""
pd.read_sql(query, engine)
```

#### 2. Average Order Value (AOV)

```python
query = """select
                round(avg(cost_of_the_order),2) as average_order_value
           from product_events;"""
pd.read_sql(query, engine)
```

#### 3. Maximum & Minimum Order Value

```python
query = """select
                max(cost_of_the_order) as highest_order,
                min(cost_of_the_order) as lowest_order
           from product_events"""
pd.read_sql(query, engine)
```

### SECTION 2: City Performance

#### 4. Revenue by City

```python
query = """select
                city,
                round(sum(cost_of_the_order),2) as revenue
           from product_events
           group by city
           order by revenue desc"""
pd.read_sql(query, engine)
```

#### 5. Average Rating by City

```python
query = """select
                city,
                round(avg(rating),2) as avg_rating
           from product_events
           group by city
           order by avg_rating desc"""
pd.read_sql(query, engine)
```

#### 6. Average Delivery Time by City

```python
query = """select
                city,
                round(avg(delivery_time),2) as avg_delivery_time
           from product_events
           group by city
           order by avg_delivery_time"""
pd.read_sql(query, engine)
```

### SECTION 3: Restaurant Performance

#### 16. Customers by Acquisition Channel

```python
query = """select
                acquisition_channel,
                count(distinct customer_id) as customers
           from product_events
           group by acquisition_channel
           order by customers desc"""
pd.read_sql(query, engine)
```

#### 18. Average Order Value by Acquisition Channel

```python
query = """select
                acquisition_channel,
                round(avg(cost_of_the_order),2) as avg_order_value
           from product_events
           group by acquisition_channel
           order by avg_order_value desc;"""
pd.read_sql(query, engine)
```

#### Revenue Contribution %

```python
query = """ select
                city,
                round(sum(cost_of_the_order),2) as revenue,
                round(
                    sum(cost_of_the_order) * 100 /
                    (select sum(cost_of_the_order) from product_events),
                     2
                     ) AS revenue_percentage
           from product_events
           group by city
           order by revenue desc"""
pd.read_sql(query, engine)
```
