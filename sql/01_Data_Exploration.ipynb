```python
query = "SHOW TABLES"
pd.read_sql(query, engine)
```

#### 1. Total Records

```python
query = """select count(*) as total_records
            from product_events"""
pd.read_sql(query, engine)
```

#### 2. Number of Customers

```python
query = """select count(distinct customer_id) as total_customers
           from product_events"""
pd.read_sql(query, engine)
```

#### 3. Number of Restaurants

```python
query = """select count(distinct restaurant_name) as total_restaurants
           from product_events"""
pd.read_sql(query, engine)
```

#### 4. Number of Cuisine Types

```python
query = """select count(distinct cuisine_type) as total_cuisines
           from product_events"""
pd.read_sql(query, engine)
```

#### 5. Date Range

```python
query = """select
                min(session_date) as first_session,
                max(session_date) as last_session
           from product_events"""
pd.read_sql(query, engine)
```

#### 6. Cities Covered

```python
query = """select distinct city
           from product_events"""
pd.read_sql(query, engine)
```

#### 7. Device Types

```python
query = """select
                device_type,
                count(*) as users
           from product_events
           group by device_type"""
pd.read_sql(query, engine)
```

#### 8. Acquisition Channels

```python
query = """select
                acquisition_channel,
                count(*) users
           from product_events
           group by acquisition_channel
           order by users desc"""
pd.read_sql(query, engine)
```

Organic acquisition channel brings the most customers

#### 9. Control vs Variant

```python
query = """select
                experiment_group,
                count(*) users
           from product_events
           group by experiment_group"""
pd.read_sql(query, engine)
```

This checks if the A/B test groups are balanced.

#### 10. Missing Values Check

```python
query = """select
                sum(customer_id is null) customer_null,
                sum(city is null) city_null,
                sum(rating is null) rating_null,
                sum(delivery_time is null) delivery_null
           from product_events"""
pd.read_sql(query, engine)
```
