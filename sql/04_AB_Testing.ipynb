```python
from sqlalchemy import create_engine

engine = create_engine(
                        f"mysql+pymysql://root:{password}@localhost:3306/foodhub"
                        )

print("Connection successful!")
```

#### 1. Users in Each Experiment Group

```python
query = """select
                experiment_group,
                count(*) as total_users
           from product_events
           group by experiment_group"""

pd.read_sql(query, engine)
```

#### 2. Recommendation Click-Through Rate (CTR)

```python
query = """select
                experiment_group,
                round(avg(recommendation_clicked) * 100, 2) as ctr
           from product_events
           group by experiment_group"""

pd.read_sql(query, engine)
```

#### 3. Cart Creation Rate

```python
query = """select
                experiment_group,
                round(avg(cart_created) * 100, 2) as cart_creation_rate
           from product_events
           group by experiment_group"""

pd.read_sql(query, engine)
```

#### 4. Average Order Value (AOV)

```python
query = """select
                experiment_group,
                round(avg(cost_of_the_order), 2) as average_order_value
           from product_events
           group by experiment_group"""

pd.read_sql(query, engine)
```

#### 5. Total Revenue by Experiment Group

```python
query = """select
                experiment_group,
                round(sum(cost_of_the_order), 2) as total_revenue
           from product_events
           group by experiment_group"""

pd.read_sql(query, engine)
```

#### 6. Average Rating

```python
query = """select
                experiment_group,
                round(avg(rating), 2) as average_rating
           from product_events
           group by experiment_group"""

pd.read_sql(query, engine)
```

#### 7. Average Delivery Time

```python
query = """select
                experiment_group,
                round(avg(delivery_time), 2) as average_delivery_time
           from product_events
           group by experiment_group"""

pd.read_sql(query, engine)
```

#### 8. Average Food Preparation Time

```python
query = """select
                experiment_group,
                round(avg(food_preparation_time), 2) as average_preparation_time
           from product_events
           group by experiment_group
"""

pd.read_sql(query, engine)
```

#### 9. Recommendation Seen %

```python
query = """select
                experiment_group,
                round(avg(recommendation_seen) * 100, 2) as recommendation_seen_rate
           from product_events
           group by experiment_group"""

pd.read_sql(query, engine)
```

#### 10. Coupon Usage %

```python
query = """select
                experiment_group,
                round(avg(coupon_used) * 100, 2) as coupon_usage_rate
           from product_events
           group by experiment_group"""

pd.read_sql(query, engine)
```

#### 11. Revenue by City and Experiment Group

```python
query = """select
                city,
                experiment_group,
                round(sum(cost_of_the_order),2) as revenue
           from product_events
           group by city, experiment_group
           order by city, revenue desc"""

pd.read_sql(query, engine)
```

#### 12. Cart Creation Rate by Device

```python
query = """select
                device_type,
                experiment_group,
                round(avg(cart_created) * 100,2) as cart_rate
           from product_events
           group by device_type, experiment_group"""

pd.read_sql(query, engine)
```

#### 13. CTR by Acquisition Channel

```python
query = """select
                acquisition_channel,
                experiment_group,
                round(avg(recommendation_clicked) * 100,2) as ctr
           from product_events
           group by acquisition_channel, experiment_group
           order by acquisition_channel"""

pd.read_sql(query, engine)
```

#### 14. Average Order Value by Device

```python
query = """select
                device_type,
                experiment_group,
                round(avg(cost_of_the_order),2) as average_order_value
           from product_events
           group by device_type, experiment_group"""

pd.read_sql(query, engine)
```

#### 15. Complete A/B Test KPI Dashboard Query

```python
query = """select
                experiment_group,
                
                count(*) as total_users,
                round(avg(recommendation_clicked) * 100,2) as ctr,
                round(avg(cart_created) * 100,2) as cart_creation_rate,

                round(avg(cost_of_the_order),2) as average_order_value,
                round(avg(delivery_time),2) as average_delivery_time,
                round(avg(rating),2) as average_rating,

                round(avg(coupon_used) * 100,2) as coupon_usage_rate

           from product_events
           group by experiment_group"""

pd.read_sql(query, engine)
```
